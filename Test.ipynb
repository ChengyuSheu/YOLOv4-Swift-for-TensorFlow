```swift
import Foundation

public func getWeights (
    _ stream: InputStream,
    _ size: Int) -> [Float] {
            var buf:[UInt8] = [UInt8](repeating: 0, count:  size * MemoryLayout<Float>.size)
            var f: [Float] = [Float](repeating: 0, count: size)
            var len = stream.read(&buf, maxLength: buf.count)
            memcpy(&f, buf, len)
            //var weights = Tensor<T>(shape:[size], scalars: f)
    return f
}

public func getWeightsInt (
    _ stream: InputStream,
    _ size: Int) -> [Int32] {
            var buf:[UInt8] = [UInt8](repeating: 0, count:  size * MemoryLayout<Int32>.size)
            var f: [Int32] = [Int32](repeating: 0, count: size)
            var len = stream.read(&buf, maxLength: buf.count)
            memcpy(&f, buf, len)
            //var weights = Tensor<T>(shape:[size], scalars: f)
    return f
}

public func load(fileAtPath path: String) -> [Float] {
        var x: [Float] = []
        if let stream:InputStream = InputStream(fileAtPath: path) {
            stream.open()
            for _ in 0..<3 {
                var s = getWeightsInt(stream, 1)
                x += getWeights(stream, Int(s[0]))
            }
            stream.close()
        }
    return x
}

```

```swift
let imageURL = URL(fileURLWithPath: "dog_512.jpg")
let img = Image(jpeg: imageURL)
var im = img.tensor.reshaped(to: [1,512,512,3]) / 255
var yolov4 = YOLOv4<Float>(cls:80).load(fileAtPath: "yolov4.weights")
yolov4.yoloHead1.test = true
yolov4.yoloHead2.test = true
yolov4.yoloHead3.test = true
```

```swift
let y = im |> yolov4
```

```swift
let x = load(fileAtPath: "output.dat")
```

```swift
var l1 = y.split(sizes: [4096, 1024, 256], alongAxis: -1)[0].flattened()
var l2 = y.split(sizes: [4096, 1024, 256], alongAxis: -1)[1].flattened()
var l3 = y.split(sizes: [4096, 1024, 256], alongAxis: -1)[2].flattened()
var m:Float = 0.0
for i in 0..<255*16*16*16 {
    let diff = abs(Float(l1[i])! - x[ i])
    m = max(m, diff)
}
print(m)

m = 0.0
for i in 0..<255*16*16*4 {
    let diff = abs(Float(l2[i])! - x[255*16*16*16 + i])
    m = max(m, diff)
}
print(m)

m = 0.0
for i in 0..<255*16*16 {
    let diff = abs(Float(l3[i])! - x[255*16*16*20 + i])
    m = max(m, diff)
}
print(m)
```
